```python
import pandas as pd
import ast
import os
```

```python
# this segmentation.csv file was created by Tan Nguyen on 08/15/2023 for all subjects, including subs38-43 
segmentation_df = pd.read_csv("/data/nil-external/dcl/Events152_fMRI_NeuralMechanisms/Analysis/segmentation_df_08_15_2023.csv")
segmentation_df["sub"].unique()
```

```python
# adapted from analyze_session_two.py
segmentation_df_long=pd.DataFrame()
i = 0
while i < len(segmentation_df):
    test=segmentation_df.iloc[i]
    if pd.notna(test['segment_fine.rt']):
        bds=ast.literal_eval(test['segment_fine.rt'])
        bds = [x for x in bds if x >= 0]
        testdf=pd.DataFrame(({'boundaries':bds}))
        testdf['grain']='fine'
        testdf['sub']=test['sub']
        testdf['order']=test['order']
        testdf['movie']=test['movie']   
        segmentation_df_long=pd.concat([segmentation_df_long,testdf],ignore_index=True)
    elif pd.notna(test['segment_coarse.rt']):
        bds = ast.literal_eval(test['segment_coarse.rt'])
        bds = [x for x in bds if x >= 0]
        testdf=pd.DataFrame(({'boundaries':bds}))
        testdf['grain']='coarse'
        testdf['sub']=test['sub']
        testdf['order']=test['order']
        testdf['movie']=test['movie'] 
        segmentation_df_long=pd.concat([segmentation_df_long,testdf],ignore_index=True)
    i+=1
segmentation_df_long
```

```python
counter_balance = pd.read_excel(r"./e152_counterbalance.xls", sheet_name='OA Participants')
counter_balance
```

```python
# convert column VC##### and fmriPrepNumber to a dictionary
vc_to_fmriprep_dict = counter_balance.set_index('VC#####').to_dict()['fmriPrep Number']
# convert column fmriPrepNumber and VC##### to a dictionary
fmriprep_to_vc_dict = counter_balance.set_index('fmriPrep Number').to_dict()['VC#####']
vc_to_fmriprep_dict
```

```python
# each row of the event file is for one run (out of 4), each file is for one participant.
# Save segmentation event files:
rundict = {'1.2.3':'run-1','6.3.9':'run-2','3.1.3':'run-3','2.4.1':'run-4'}
vc_subs = segmentation_df_long['sub'].unique()
for vc_sub in vc_subs:
    # check if this segmentation participant has a fMRI scan (to exlude pilot, etc.)
    if not (isinstance(vc_to_fmriprep_dict[vc_sub], str) and "sub" in vc_to_fmriprep_dict[vc_sub]):
        continue
    if not os.path.exists("EVTs"):
        os.makedirs("EVTs")
    if not os.path.exists(f"EVTs/{vc_to_fmriprep_dict[vc_sub]}/"):
        os.makedirs(f"EVTs/{vc_to_fmriprep_dict[vc_sub]}/")
    else:
        print(f'participant {vc_sub} already has segmentation event files, continue...')
        continue
    for grain in ['fine', 'coarse']:
        for movie in ['1.2.3','6.3.9','3.1.3','2.4.1']:
            try:
                test = segmentation_df_long[(segmentation_df_long['sub']==vc_sub) & (segmentation_df_long['movie']==movie) & (segmentation_df_long['grain'] == grain)].copy()
                event_boundaries = test['boundaries'].tolist()
                event_boundaries = [str(round(x, 2)) for x in event_boundaries]
                with open(f"EVTs/{vc_to_fmriprep_dict[vc_sub]}/" + vc_to_fmriprep_dict[vc_sub] + f'_movie_{grain}.txt', 'at') as f:
                    # if the participant doesn't have segmentation for this run and grain, write * (afni convention)
                    if len(event_boundaries) == 0:
                        f.write("*\n")
                    else:
                        f.write(" ".join(event_boundaries) + "\n")
            except Exception as e:
                print(e, vc_sub, movie, test)
```

```python
segmentation_df['fmriprep'] = segmentation_df['sub'].map(vc_to_fmriprep_dict)
segmentation_df.groupby(["sub", "fmriprep"]).count()
```
